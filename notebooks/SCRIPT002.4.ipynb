## Function basic

### Function Basic Sample

````python
def Simple(a):
    print('a = {}'.format(a))

def ChangeArg(a):
    a += 2
    print('a = {}'.format(a))
    
def ChangeList(l):
    l.append(999)
    print('l = {}'.format(l))

    
x = 4
Simple(x)
print('x = {}'.format(x))
ChangeArg(x)
print('x = {}'.format(x))
k = [1,2,3,4]
print('k = {}'.format(k))
ChangeList(k)
print('k = {}'.format(k))
````

### Functions Arguments sample

````python
def MakeIt(a,b,c=9,d=7):
    print('a = {} b = {} c = {} d = {}'.format(a,b,c,d))
    

def sum1(a,*args):
    s = a
    for i in args:
        s += i
    return s

def sum2(a,**keyargs):
    s = a
    for i in keyargs:
        s += i
    return s

def sum3(a,*args,**keyargs):
    s = a
    for i in args:
        s += i
    for k,i in keyargs.items():
        s += i
    return s

MakeIt(1,2)
MakeIt(1,2,3)
MakeIt(1,2,3,4)
print(sum1(9,2,3,4,5))
print(sum2(9,m = 2,n = 3,k = 4,z = 5))
print(sum3(9,1,2,4,m = 2,n = 3,k = 4,z = 5))
````

Change function *sum2()* for calculate sum int values. What we change so that operator **for** work with dictionary items, not keys?

### Fibonacci  Sample


````python
current = 1
next = 1
def fibonacci():
    global current
    global next
    prev =  current
    current = next
    next = prev + current
    return prev

for i in range(10):
    print(fibonacci())
````

### Nonlocal sample

````python
x = 3
def MakeIt(a):
    x = a + 2;
    def Internal():
        nonlocal x
        return x + 2
    return Internal()
print(MakeIt(5))
````

## Lambda Function

### Lambda Simple sample

````python
def printList(l,p):
    for i in l:
        if p(i):
            print(i)

printList([-2,5,0,-7,-6,5,4],lambda x: x > 0)
````

### Lambda factorial

````python
Fact = lambda x : Fact(x-1)*x if x > 1 else 1
print(Fact(5))
````
